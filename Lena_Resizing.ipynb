```python
import cv2
import requests
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# URL da imagem da Lena
url_imagem = "https://upload.wikimedia.org/wikipedia/en/7/7d/Lenna_%28test_image%29.png"

# Baixar a imagem da URL
response = requests.get(url_imagem)
imagem_array = np.asarray(bytearray(response.content), dtype=np.uint8)
imagem_colorida = cv2.imdecode(imagem_array, cv2.IMREAD_COLOR)

# Converter para escala de cinza
imagem_cinza = cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2GRAY)

# Binarizar a imagem
_, imagem_binarizada = cv2.threshold(imagem_cinza, 128, 255, cv2.THRESH_BINARY)

# Exibir as imagens lado a lado
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(imagem_colorida, cv2.COLOR_BGR2RGB))
plt.title('Figura 1: Lena colorida')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(imagem_cinza, cmap='gray', vmin=0, vmax=255)
plt.title('NÃ­veis de Cinza')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(imagem_binarizada, cmap='gray', vmin=0, vmax=255)
plt.title('Binarizada')
plt.axis('off')

plt.show()
```

# Report: Color Image to Grayscale and Binarization Transformation

## Introduction

This report addresses the implementation of a Python code to transform a color image into grayscale and subsequently binarize it. The code utilizes the OpenCV library for image processing and matplotlib for displaying the results. The chosen image for this study is the famous "Lena," often used in image processing as a benchmark.

## Methods

The developed code follows the following steps:

1. **Image Loading:** The Lena image is loaded, either locally or from a URL on the internet, using the `requests` library for the latter case.

2. **Conversion to Grayscale:** The color image is converted to grayscale using the `cv2.cvtColor` function from the OpenCV library.

3. **Image Binarization:** The grayscale image is binarized using the `cv2.threshold` function, which applies a threshold to transform pixels into black or white.

4. **Displaying Results:** The three images (color, grayscale, and binarized) are displayed side by side to allow for visual comparison.

## Results

The obtained results clearly demonstrate the transformations applied to the Lena image. The color image is shown alongside the grayscale and binarized versions. The conversion to grayscale highlights the luminance information of the image, while binarization simplifies the image to only two intensity levels, black and white.
