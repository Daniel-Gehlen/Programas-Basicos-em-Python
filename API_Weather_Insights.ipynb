```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Função para obter os dados meteorológicos com base na latitude e longitude
def obter_dados_meteorologicos(lat, lon, api_key):
    url = f'https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={api_key}&units=metric'
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print("Falha ao obter os dados. Verifique sua chave de API ou sua conexão com a internet.")

# Substitua 'sua_chave_de_api' pela sua própria chave de API
api_key = 'sua_chave_de_api'

# Dicionário com as capitais dos estados do Brasil e suas respectivas coordenadas (latitude e longitude)
capitais_br = {
    'AC': (-9.974722, -67.824722),
    'AL': (-9.666667, -35.735),
    'AP': (0.035, -51.05),
    'AM': (-3.10194, -60.025),
    'BA': (-12.971389, -38.501389),
    'CE': (-3.71722, -38.54306),
    'DF': (-15.793889, -47.882778),
    'ES': (-20.315278, -40.3125),
    'GO': (-16.678056, -49.253889),
    'MA': (-2.538611, -44.282222),
    'MT': (-15.601667, -56.0975),
    'MS': (-20.442778, -54.646389),
    'MG': (-19.916667, -43.933333),
    'PA': (-1.455833, -48.504444),
    'PB': (-7.115, -34.863056),
    'PR': (-25.4284, -49.2733),
    'PE': (-8.047778, -34.877778),
    'PI': (-5.09194, -42.803889),
    'RJ': (-22.908333, -43.196389),
    'RN': (-5.794444, -35.211667),
    'RS': (-30.032778, -51.230000),
    'RO': (-8.761944, -63.903889),
    'RR': (2.819722, -60.673056),
    'SC': (-27.5956, -48.548),
    'SP': (-23.550278, -46.633333),
    'SE': (-10.947222, -37.073056),
    'TO': (-10.175278, -48.298056)
}

# Lista para armazenar os dados meteorológicos de cada capital
dados_capitais = []

# Obtendo os dados meteorológicos de cada capital
for estado, coordenadas in capitais_br.items():
    lat, lon = coordenadas
    dados = obter_dados_meteorologicos(lat, lon, api_key)
    dados_capitais.append(dados)

# Convertendo os dados em um DataFrame do Pandas
df = pd.DataFrame(dados_capitais)

# Exibindo as primeiras linhas do DataFrame
display(df.head())

# Gerando uma tabela com algumas estatísticas comuns
tabela_estatisticas = df.describe()
display(tabela_estatisticas)

# Plotando um gráfico de barras da temperatura média por estado
plt.figure(figsize=(12, 6))
plt.bar(df['name'], df['main'].apply(lambda x: x['temp']).astype(float), color='skyblue')
plt.title('Temperatura Média por Capital')
plt.xlabel('Capital')
plt.ylabel('Temperatura Média (°C)')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Introduction:

The BrazilWeather Insights project aims to explore and analyze meteorological data from the capitals of Brazilian states. Using the OpenWeatherMap API, this project collects information about temperature, humidity, wind speed, and other meteorological variables for each capital of Brazil. The goal is to provide valuable insights into the climate in different regions of the country and demonstrate how this data can be visualized and interpreted.

---

## Methods:

To carry out this project, the following steps were taken:

1. **Data Acquisition**: The OpenWeatherMap API was used to obtain meteorological data for each capital of Brazil based on their latitude and longitude coordinates.

2. **Data Analysis**: The obtained data was processed and organized into a DataFrame using the Pandas library. Descriptive statistical analyses and graphical visualizations were performed to understand trends and patterns in the meteorological data.

3. **Data Visualization**: The data was visualized through bar charts, allowing for comparison of average temperatures between the capitals.

---

## Results:

After analyzing the data, the following results were obtained:

- Capitals located in the North and Northeast regions of Brazil tend to have higher temperatures compared to those in the South and Southeast.
- Relative humidity varies significantly between capitals, reflecting differences in climatic conditions in each region.
- Wind speed also varies between capitals, with some experiencing stronger winds than others.
